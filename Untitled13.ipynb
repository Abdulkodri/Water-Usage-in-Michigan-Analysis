## Analysis of Water Usage in Michigan by Industry and Source (2013-2022)

This project focuses on analyzing water usage trends across various industries in Michigan from 2013 to 2022. The data, sourced from public records on Michigan.gov, provides a comprehensive overview of water withdrawals categorized by three main sources: the Great Lakes, groundwater, and inland surface water.

By examining the volume of water withdrawn by different industry sectors—such as commercial-institutional, industrial-manufacturing, and irrigation—this analysis aims to identify trends and patterns in water usage. The key variables in the dataset include the industry, year, and the amount of water withdrawn from each source, with total gallons calculated as the sum of withdrawals from all sources.

Potential uses of this analysis include understanding industry-specific water dependencies, assessing the sustainability of water use practices, and informing conservation strategies. By comparing withdrawal patterns from the Great Lakes, groundwater, and inland surface water, the project seeks to provide valuable insights into water resource management and policy-making in Michigan.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Load the dataset
data = pd.read_csv('water_use_data_2013_to_2022.csv')
data
```

```python
# Display the first few rows of the dataset
print(data.head())3
```

```python
# Display dataset information
print(data.info())
```

```python
# Display summary statistics
print(data.describe())
```

```python
# Checking for missing values
print(data.isnull().sum())
```

```python
data.columns = data.columns.str.strip()
```

```python
# Verify the cleaned column names
print(data.columns)
```

```python
# Rename columns if necessary (adjust the names as per your dataset)
data.rename(columns={
    'Great Lakes Gallons': 'Great Lakes (Gallons)',
    'Groundwater Gallons': 'Groundwater (Gallons)',
    'Inland Surface Water Gallons': 'Inland Surface Water (Gallons)'
}, inplace=True)
```

```python
# Verify the renamed column names
print(data.columns)
```

```python
# Add a new column for total water usage
data['Total Gallons'] = data['gallons_from_great_lakes'] + data['gallons_from_groundwater'] + data['gallons_from_inland_surface']
print(data.head())
```

```python
# Exploratory Data Analysis (EDA)
# Distribution of total water usage by year
plt.figure(figsize=(12, 6))
sns.barplot(x='year', y='Total Gallons', data=data, ci=None)
plt.title("Total Water Usage by Year")
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of water usage by industry
plt.figure(figsize=(12, 6))
sns.barplot(x='industry', y='Total Gallons', data=data, ci=None)
plt.title("Total Water Usage by Industry")
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribution of water sources by industry
plt.figure(figsize=(12, 6))
data.groupby('industry').sum()[['gallons_from_great_lakes', 'gallons_from_groundwater', 'gallons_from_inland_surface']].plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title("Water Sources by Industry")
plt.xlabel("Industry")
plt.ylabel("Gallons")
plt.xticks(rotation=90)
plt.show()
```

```python
# Water usage trends over time for each industry
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='Total Gallons', hue='industry', data=data)
plt.title("Water Usage Trends Over Time by Industry")
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation analysis between different water sources
plt.figure(figsize=(12, 6))
sns.heatmap(data[['gallons_from_great_lakes', 'gallons_from_groundwater', 'gallons_from_inland_surface', 'Total Gallons']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Summary and Conclusions
total_usage_by_year = data.groupby('year')['Total Gallons'].sum()
total_usage_by_industry = data.groupby('industry')['Total Gallons'].sum()
great_lakes_usage = data['gallons_from_great_lakes'].sum()
groundwater_usage = data['gallons_from_groundwater'].sum()
inland_surface_water_usage = data['inland_surface_water_usage (gallons)'].sum()
```

```python
print(f"Total water usage by year:\n{total_usage_by_year}\n")
print(f"Total water usage by industry:\n{total_usage_by_industry}\n")
print(f"Total Great Lakes water usage: {great_lakes_usage} gallons")
print(f"Total groundwater usage: {groundwater_usage} gallons")
print(f"Total inland surface water usage: {inland_surface_water_usage} gallons")
```
